```python
# namedtuple 생성
from collections import namedtuple, OrderedDict

Transition = namedtuple(
    'Transition', ('state', 'action', 'next_state', 'reward'))
```

```python
# transition을 저장하기 위한 메모리 클래스


class ReplayMemory:

    def __init__(self, CAPACITY):
        self.capacity = CAPACITY  # 메모리의 최대 저장 건수
        self.memory = []  # 실제 transition을 저장할 변수
        self.index = 0  # 저장 위치를 가리킬 인덱스 변수

    def push(self, state, action, state_next, reward):
        '''transition = (state, action, state_next, reward)을 메모리에 저장'''

        if len(self.memory) < self.capacity:
            self.memory.append(None)  # 메모리가 가득차지 않은 경우

        # Transition이라는 namedtuple을 사용하여 키-값 쌍의 형태로 값을 저장
        self.memory[self.index] = Transition(state, action, state_next, reward)

        self.index = (self.index + 1) % self.capacity  # 다음 저장할 위치를 한 자리 뒤로 수정

    def sample(self, batch_size):
        '''batch_size 갯수 만큼 무작위로 저장된 transition을 추출'''
        return random.sample(self.memory[:-MAX_STEPS], batch_size)

    def __len__(self):
        '''len 함수로 현재 저장된 transition 갯수를 반환'''
        return len(self.memory)
```

```python
# Dueling Network 신경망 구성
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):

    def __init__(self, n_in, n_mid, n_out):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(n_in, n_mid)
        self.fc2 = nn.Linear(n_mid, n_mid)
        self.fc3 = nn.Linear(n_mid, n_mid)
        # Dueling Network
        self.fc4_adv = nn.Linear(n_mid, n_out)  # Advantage함수쪽 신경망
        self.fc4_v = nn.Linear(n_mid, 1)  # 가치 V쪽 신경망

    def forward(self, x):
        h1 = F.relu(self.fc1(x))
        h2 = F.relu(self.fc2(h1))
        h3 = F.relu(self.fc3(h2))

        adv = self.fc4_adv(h3)  # 이 출력은 ReLU를 거치지 않음
        val = self.fc4_v(h3).expand(-1, adv.size(1))  # 이 출력은 ReLU를 거치지 않음
        # val은 adv와 덧셈을 하기 위해 expand 메서드로 크기를 [minibatch*1]에서 [minibatch*2]로 변환
        # adv.size(1)은 2(출력할 행동의 가짓수)

        output = val + adv - adv.mean(1, keepdim=True).expand(-1, adv.size(1))
        # val+adv에서 adv의 평균을 뺀다
        # adv.mean(1, keepdim=True) 으로 열방향(행동의 종류 방향) 평균을 구함 크기는 [minibatch*1]이 됨
        # expand 메서드로 크기를 [minibatch*2]로 늘림
        return output
```

```python
# 에이전트의 두뇌 역할을 하는 클래스, DDQN을 실제 수행한다 

import random
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F

BATCH_SIZE = 128
CAPACITY = 10000


class Brain:
    def __init__(self, num_states, num_actions):
        self.num_actions = num_actions 

        # transition을 기억하기 위한 메모리 객체 생성
        self.memory = ReplayMemory(CAPACITY)

        # 신경망 구성
        n_in, n_mid, n_out = num_states, 32, num_actions
        self.main_q_network = Net(n_in, n_mid, n_out).cuda()  # Net 클래스를 사용
        self.target_q_network = Net(n_in, n_mid, n_out).cuda()  # Net 클래스를 사용
#         print(self.main_q_network)  # 신경망의 구조를 출력

        # 최적화 기법 선택
        self.optimizer = optim.Adam(
            self.main_q_network.parameters(), lr=0.00005)

    def replay(self):
        '''Experience Replay로 신경망의 결합 가중치 학습'''

        # 1. 저장된 transition의 수를 확인
        if len(self.memory) < 2*BATCH_SIZE:
            return

        # 2. 미니배치 생성
        self.batch, self.state_batch, self.action_batch, self.reward_batch, self.non_final_next_states = self.make_minibatch()        
        
        # 3. 정답신호로 사용할 Q(s_t, a_t)를 계산
        self.expected_state_action_values = self.get_expected_state_action_values()

        # 4. 결합 가중치 수정
        self.update_main_q_network()

    def decide_action(self, state, episode, istest):
        '''현재 상태로부터 행동을 결정함'''
        # ε-greedy 알고리즘에서 서서히 최적행동의 비중을 늘린다
        epsilon = 0.5 * (1 / (episode + 1))

        if epsilon <= np.random.uniform(0, 1) or istest == True:
            self.main_q_network.eval()  # 신경망을 추론 모드로 전환
            with torch.no_grad():
                action = self.main_q_network(state).max(1)[1].view(1, 1)
            # 신경망 출력의 최댓값에 대한 인덱스 = max(1)[1]
            # .view(1,1)은 [torch.LongTensor of size 1] 을 size 1*1로 변환하는 역할을 한다

        else:
            # 행동을 무작위로 반환(0 혹은 1)
            action = torch.LongTensor(
                [[random.randrange(self.num_actions)]])  # 행동을 무작위로 반환
            # action은 [torch.LongTensor of size 1*1] 형태가 된다 

        return action

    def make_minibatch(self):
        '''2. 미니배치 생성'''

        # 2.1 메모리 객체에서 미니배치를 추출
        transitions = self.memory.sample(BATCH_SIZE)

        # 2.2 각 변수를 미니배치에 맞는 형태로 변형
        # transitions는 각 단계 별로 (state, action, state_next, reward) 형태로 BATCH_SIZE 갯수만큼 저장됨
        # 다시 말해, (state, action, state_next, reward) * BATCH_SIZE 형태가 된다
        # 이것을 미니배치로 만들기 위해
        # (state*BATCH_SIZE, action*BATCH_SIZE, state_next*BATCH_SIZE, reward*BATCH_SIZE) 형태로 변환한다
        batch = Transition(*zip(*transitions))

        # 2.3 각 변수의 요소를 미니배치에 맞게 변형하고, 신경망으로 다룰 수 있도록 Variable로 만든다
        # state를 예로 들면, [torch.FloatTensor of size 1*4] 형태의 요소가 BATCH_SIZE 갯수만큼 있는 형태이다
        # 이를 torch.FloatTensor of size BATCH_SIZE*4 형태로 변형한다
        # 상태, 행동, 보상, non_final 상태로 된 미니배치를 나타내는 Variable을 생성
        # cat은 Concatenates(연접)을 의미한다

        state_batch = torch.cat(batch.state).to(device)
        action_batch = torch.cat(batch.action).to(device)
        reward_batch = torch.cat(batch.reward).to(device)
        non_final_next_states = torch.cat([s for s in batch.next_state
                                           if s is not None]).to(device)

        return batch, state_batch, action_batch, reward_batch, non_final_next_states

    def get_expected_state_action_values(self):
        '''정답신호로 사용할 Q(s_t, a_t)를 계산'''

        # 3.1 신경망을 추론 모드로 전환
        self.main_q_network.eval()
        self.target_q_network.eval()

        # 3.2 신경망으로 Q(s_t, a_t)를 계산
        # self.model(state_batch)은 왼쪽, 오른쪽에 대한 Q값을 출력하며
        # [torch.FloatTensor of size BATCH_SIZEx2] 형태이다
        # 여기서부터는 실행한 행동 a_t에 대한 Q값을 계산하므로 action_batch에서 취한 행동 a_t가 
        # 왼쪽이냐 오른쪽이냐에 대한 인덱스를 구하고, 이에 대한 Q값을 gather 메서드로 모아온다
        
        self.state_action_values = self.main_q_network(
            self.state_batch).gather(1, self.action_batch)
        
        
        # 3.3 max{Q(s_t+1, a)}값을 계산한다 이때 다음 상태가 존재하는지에 주의해야 한다

        # done 상태가 아니고, next_state가 존재하는지 확인하는 인덱스 마스크를 만듬
        non_final_mask = torch.ByteTensor(tuple(map(lambda s: s is not None,
                                                    self.batch.next_state)))
        # 먼저 전체를 0으로 초기화
        next_state_values = torch.zeros(BATCH_SIZE).cuda()

        a_m = torch.zeros(BATCH_SIZE).type(torch.LongTensor).cuda()

        # 다음 상태에서 Q값이 최대가 되는 행동 a_m을 Main Q-Network로 계산
        # 마지막에 붙은 [1]로 행동에 해당하는 인덱스를 구함
        
        
        a_m[non_final_mask] = self.main_q_network(
            self.non_final_next_states).detach().max(1)[1]
        
        # 다음 상태가 있는 것만을 걸러내고, size 32를 32*1로 변환
        a_m_non_final_next_states = a_m[non_final_mask].view(-1, 1)

        # 다음 상태가 있는 인덱스에 대해 행동 a_m의 Q값을 target Q-Network로 계산
        # detach() 메서드로 값을 꺼내옴
        # squeeze() 메서드로 size[minibatch*1]을 [minibatch]로 변환
        next_state_values[non_final_mask] = self.target_q_network(
            self.non_final_next_states).gather(1, a_m_non_final_next_states).detach().squeeze()

        
        # 3.4 정답신호로 사용할 Q(s_t, a_t)값을 Q러닝 식으로 계산한다
        expected_state_action_values = self.reward_batch + GAMMA * next_state_values
        
        return expected_state_action_values

    def update_main_q_network(self):
        '''4. 결합 가중치 수정'''

        # 4.1 신경망을 학습 모드로 전환
        self.main_q_network.train()

        # 4.2 손실함수를 계산 (smooth_l1_loss는 Huber 함수)
        # expected_state_action_values은
        # size가 [minibatch]이므로 unsqueeze하여 [minibatch*1]로 만든다
        loss = F.smooth_l1_loss(self.state_action_values,
                                self.expected_state_action_values.unsqueeze(1))

        # 4.3 결합 가중치를 수정한다
        self.optimizer.zero_grad()  # 경사를 초기화
        loss.backward()  # 역전파 계산
        self.optimizer.step()  # 결합 가중치 수정

    def update_target_q_network(self):  # DDQN에서 추가됨
        '''Target Q-Network을 Main Q-Network와 맞춤'''
        self.target_q_network.load_state_dict(self.main_q_network.state_dict())
```

```python
# 에이전트 클래스

class Agent:
    def __init__(self, num_states, num_actions):
        '''태스크의 상태 및 행동의 가짓수를 설정'''
        self.brain = Brain(num_states, num_actions)  # 에이전트의 행동을 결정할 두뇌 역할 객체를 생성

    def update_q_function(self):
        '''Q함수를 수정'''
        self.brain.replay()

    def get_action(self, state, episode, istest=False):
        '''행동을 결정'''
        action = self.brain.decide_action(state, episode, istest)
        return action

    def memorize(self, state, action, state_next, reward):
        '''memory 객체에 state, action, state_next, reward 내용을 저장'''
        self.brain.memory.push(state, action, state_next, reward)

    def update_target_q_function(self):
        '''Target Q-Network을 Main Q-Network와 맞춤'''
        self.brain.update_target_q_network()
        
```

```python
# 실행 환경 클래스

class Environment:

    def __init__(self, env_type, BS_level):
        self.env = gym.make('BeerGame-v0', n_agents=4, env_type=env_type, BS_level=BS_level, n_turns_per_game=MAX_STEPS)
#         print(env_type)
        num_states = len(torch.tensor(self.env.reset()).type(torch.FloatTensor)[:,:5].reshape(-1))
        num_alphas = 11
        num_actions = 11 * num_alphas
        
        tmp = []
        for each in [[i]*11 for i in range(-5,6)] :
            tmp += each
            
        self.d_x_rule_dict = {k:tmp[k] for k in range(len(tmp))} 
        self.alpha_dict = {k:i for k,i in enumerate([i/10 for i in range(11)]*11)}
        
        self.BS_level = BS_level
        self.agent_0 = Agent(num_states, num_actions)  # retailer 역할을 할 객체를 생성
        self.agent_1 = Agent(num_states, num_actions)  # wholesaler 역할을 할 객체를 생성
        self.agent_2 = Agent(num_states, num_actions)  # distributor 역할을 할 객체를 생성
        self.agent_3 = Agent(num_states, num_actions)  # manufacturer 역할을 할 객체를 생성
        
    def train(self):

        '''실행'''

        rewards_avg_list_ret = []
        rewards_avg_list_who = []
        rewards_avg_list_dis = []
        rewards_avg_list_man = []

        for episode in tqdm(range(NUM_EPISODES)):  # 최대 에피소드 수만큼 반복
        
        
            observation = self.env.reset()  # 환경 초기화
            rewards_list = []
            state = observation  # 관측을 변환없이 그대로 상태 s로 사용

            state = torch.tensor(state).type(torch.FloatTensor).cuda()


            for step in range(MAX_STEPS):  # 1 에피소드에 해당하는 반복문
                
                
                state_ret = state[:,:5].reshape(-1).unsqueeze(0)
                state_who = state[:,5:10].reshape(-1).unsqueeze(0)
                state_dis = state[:,10:15].reshape(-1).unsqueeze(0)
                state_man = state[:,15:].reshape(-1).unsqueeze(0)
                
#                 action = self.env._get_BS_policy()

                action_ret = self.agent_0.get_action(state_ret, episode, istest=False).cuda()  # 다음 행동을 결정
                action_ret_ = self.d_x_rule_dict[action_ret.item()] # d + x 를 위한 변환
                action_who = self.agent_1.get_action(state_who, episode, istest=False).cuda()  # 다음 행동을 결정
                action_who_ = self.d_x_rule_dict[action_who.item()] # d + x 를 위한 변환
                action_dis = self.agent_2.get_action(state_dis, episode, istest=False).cuda()  # 다음 행동을 결정
                action_dis_ = self.d_x_rule_dict[action_dis.item()] # d + x 를 위한 변환
                action_man = self.agent_3.get_action(state_man, episode, istest=False).cuda()  # 다음 행동을 결정
                action_man_ = self.d_x_rule_dict[action_man.item()] # d + x 를 위한 변환

#                 action[3] = action_man_
                action = [action_ret_, action_who_, action_dis_, action_man_]

    
                observation_next, reward, done, info = self.env.step(action) # action은 주문량 d + x rule의 x에 해당됨
                
                if episode in [1, 300, NUM_EPISODES - 1] :
                    self.env.render(episode, './실험결과/{}/log.txt'.format(PATH))

                # 보상을 부여하고 episode의 종료 판정 및 state_next를 설정한다
                if done:
                    state_next = None  # 다음 상태가 없으므로 None으로
                    break

                else:
                    state_next = observation_next  # 관측 결과를 그대로 상태로 사용
                    state_next = torch.tensor(state_next).type(torch.FloatTensor).cuda()
                    state_ret_next = state_next[:,:5].reshape(-1).unsqueeze(0)
                    state_who_next = state_next[:,5:10].reshape(-1).unsqueeze(0)
                    state_dis_next = state_next[:,10:15].reshape(-1).unsqueeze(0)
                    state_man_next = state_next[:,15:].reshape(-1).unsqueeze(0)

                # 메모리에 경험을 저장
                self.agent_0.memorize(state_ret, action_ret, state_ret_next, torch.tensor(reward[0]).unsqueeze(0))
                self.agent_1.memorize(state_who, action_who, state_who_next, torch.tensor(reward[1]).unsqueeze(0))
                self.agent_2.memorize(state_dis, action_dis, state_dis_next, torch.tensor(reward[2]).unsqueeze(0))
                self.agent_3.memorize(state_man, action_man, state_man_next, torch.tensor(reward[3]).unsqueeze(0))

                # Experience Replay로 Q함수를 수정
                self.agent_0.update_q_function()
                self.agent_1.update_q_function()
                self.agent_2.update_q_function()
                self.agent_3.update_q_function()

                # 관측 결과를 업데이트
                state = state_next
                rewards_list.append(reward)
                
            
            total_avg_reward = np.mean(rewards_list) 
            rewards_avg_ret = np.mean(np.array(rewards_list)[:,:1])
            rewards_avg_who = np.mean(np.array(rewards_list)[:,1:2])
            rewards_avg_dis = np.mean(np.array(rewards_list)[:,2:3])
            rewards_avg_man = np.mean(np.array(rewards_list)[:,3:])
            
            rewards_avg_steps =[]
            
            for step in zip(np.array(rewards_list)[:,:1],np.array(rewards_list)[:,1:2],np.array(rewards_list)[:,2:3],np.array(rewards_list)[:,3:4]) :
                rewards_avg_steps.append(np.mean(step))
                
                
            
            for i, step in enumerate(self.agent_0.brain.memory.memory[-MAX_STEPS+1:]) :
                ALPHA = self.alpha_dict[step.action[0].item()]
                step.reward[0] = step.reward[0] * (1-ALPHA) + ALPHA * rewards_avg_steps[i]
                
            for i, step in enumerate(self.agent_1.brain.memory.memory[-MAX_STEPS+1:]) :
                ALPHA = self.alpha_dict[step.action[0].item()]
                step.reward[0] = step.reward[0] * (1-ALPHA) + ALPHA * rewards_avg_steps[i]
                
            for i, step in enumerate(self.agent_2.brain.memory.memory[-MAX_STEPS+1:]) :
                ALPHA = self.alpha_dict[step.action[0].item()]
                step.reward[0] = step.reward[0] * (1-ALPHA) + ALPHA * rewards_avg_steps[i]
                
            for i, step in enumerate(self.agent_3.brain.memory.memory[-MAX_STEPS+1:]) :
                ALPHA = self.alpha_dict[step.action[0].item()]
                step.reward[0] = step.reward[0] * (1-ALPHA) + ALPHA * rewards_avg_steps[i]
          
            rewards_avg_list_ret.append(rewards_avg_ret)
            rewards_avg_list_who.append(rewards_avg_who)
            rewards_avg_list_dis.append(rewards_avg_dis)
            rewards_avg_list_man.append(rewards_avg_man)


            # DDQN으로 추가된 부분 5에피소드마다 한번씩 Target Q-Network을 Main Q-Network와 맞춰줌
            if(episode % 5 == 0):
                self.agent_0.update_target_q_function()
                self.agent_1.update_target_q_function()
                self.agent_2.update_target_q_function()
                self.agent_3.update_target_q_function()
        
            if episode in [500, 600, 700, 800, NUM_EPISODES - 1] :
                torch.save(self.agent_0.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_ret_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_1.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_who_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_2.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_dis_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_3.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_man_EPISODE {:04}.pt'.format(PATH, episode))
                
                torch.save(self.agent_0.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_ret_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_1.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_who_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_2.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_dis_EPISODE {:04}.pt'.format(PATH, episode))
                torch.save(self.agent_3.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_man_EPISODE {:04}.pt'.format(PATH, episode))
            
                
                
            if episode % 50 == 0 or episode == (NUM_EPISODES - 1):
                plot(rewards_avg_list_ret, rewards_avg_list_who, rewards_avg_list_dis, rewards_avg_list_man)
                with open("./실험결과/{}/학습 정보.txt".format(PATH), "a") as f:
                    f.write("\n\n")
                    f.write("-------- EPISODE {}--------\n".format(episode))
                    f.write("현재 ret 평균 리워드 : {}\n".format(rewards_avg_ret))
                    f.write("현재 who 평균 리워드 : {}\n".format(rewards_avg_who))
                    f.write("현재 dis 평균 리워드 : {}\n".format(rewards_avg_dis))
                    f.write("현재 man 평균 리워드 : {}\n".format(rewards_avg_man))
                    f.write("현재 ALL 평균 리워드 : {}\n".format(total_avg_reward))
                    f.write("\n\n")
                    
            if episode == 0 :
                self.max_avg_reward = total_avg_reward
            elif episode > 300 and total_avg_reward > self.max_avg_reward :
                
                with open("./실험결과/{}/Max Rewards 정보.txt".format(PATH), "w") as f:
                    f.write("-------- EPISODE {}--------\n".format(episode))
                    f.write("현재 ret 평균 리워드 : {}\n".format(rewards_avg_ret))
                    f.write("현재 who 평균 리워드 : {}\n".format(rewards_avg_who))
                    f.write("현재 dis 평균 리워드 : {}\n".format(rewards_avg_dis))
                    f.write("현재 man 평균 리워드 : {}\n".format(rewards_avg_man))
                    f.write("현재 ALL 평균 리워드 : {}\n".format(total_avg_reward))
                    f.write("\n\n")
                
                self.max_avg_reward = total_avg_reward
                torch.save(self.agent_0.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_ret_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_1.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_who_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_2.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_dis_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_3.brain.main_q_network.state_dict(), './실험결과/{}/main_q_net_man_EPISODE {:04}.pt'.format(PATH, 0))
                
                torch.save(self.agent_0.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_ret_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_1.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_who_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_2.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_dis_EPISODE {:04}.pt'.format(PATH, 0))
                torch.save(self.agent_3.brain.target_q_network.state_dict(), './실험결과/{}/targ_q_net_man_EPISODE {:04}.pt'.format(PATH, 0))

        return rewards_avg_list_ret, rewards_avg_list_who, rewards_avg_list_dis, rewards_avg_list_man
    
    def test(self):
        
        
        for episode in [0, 500, NUM_EPISODES - 1]:  # 최대 에피소드 수만큼 반복

            self.agent_0.brain.main_q_network.load_state_dict(torch.load('./실험결과/{}/main_q_net_ret_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_1.brain.main_q_network.load_state_dict(torch.load('./실험결과/{}/main_q_net_who_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_2.brain.main_q_network.load_state_dict(torch.load('./실험결과/{}/main_q_net_dis_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_3.brain.main_q_network.load_state_dict(torch.load('./실험결과/{}/main_q_net_man_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_0.brain.target_q_network.load_state_dict(torch.load('./실험결과/{}/targ_q_net_ret_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_1.brain.target_q_network.load_state_dict(torch.load('./실험결과/{}/targ_q_net_who_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_2.brain.target_q_network.load_state_dict(torch.load('./실험결과/{}/targ_q_net_dis_EPISODE {:04}.pt'.format(PATH, episode)))
            self.agent_3.brain.target_q_network.load_state_dict(torch.load('./실험결과/{}/targ_q_net_man_EPISODE {:04}.pt'.format(PATH, episode)))
            
            
            stock_list_ret = []
            stock_list_who = []
            stock_list_dis = []
            stock_list_man = []

            demand_list_ret = []
            demand_list_who = []
            demand_list_dis = []
            demand_list_man = []

            action_real_list_ret = []
            action_real_list_who = []
            action_real_list_dis = []
            action_real_list_man = []

            rewards_list_ret = []
            rewards_list_who = []
            rewards_list_dis = []
            rewards_list_man = []

            
            observation = self.env.reset()  # 환경 초기화
            rewards_list = []
            state = observation  # 관측을 변환없이 그대로 상태 s로 사용

            state = torch.tensor(state).type(torch.FloatTensor).cuda()


            for step in range(MAX_STEPS):  # 1 에피소드에 해당하는 반복문
                                
                state_ret = state[:,:5].reshape(-1).unsqueeze(0)
                state_who = state[:,5:10].reshape(-1).unsqueeze(0)
                state_dis = state[:,10:15].reshape(-1).unsqueeze(0)
                state_man = state[:,15:].reshape(-1).unsqueeze(0)
                
#                 action = self.env._get_BS_policy()
                
                action_ret = self.agent_0.get_action(state_ret, episode, istest=True).cuda()  # 다음 행동을 결정
                action_ret_ = self.d_x_rule_dict[action_ret.item()] # d + x 를 위한 변환
                action_who = self.agent_1.get_action(state_who, episode, istest=True).cuda()  # 다음 행동을 결정
                action_who_ = self.d_x_rule_dict[action_who.item()] # d + x 를 위한 변환
                action_dis = self.agent_2.get_action(state_dis, episode, istest=True).cuda()  # 다음 행동을 결정
                action_dis_ = self.d_x_rule_dict[action_dis.item()] # d + x 를 위한 변환
                action_man = self.agent_3.get_action(state_man, episode, istest=True).cuda()  # 다음 행동을 결정
                action_man_ = self.d_x_rule_dict[action_man.item()] # d + x 를 위한 변환
#                 action[3] = action_man_
                action = [action_ret_, action_who_, action_dis_, action_man_]

                observation_next, reward, done, info = self.env.step(action) # action은 주문량 d + x rule의 x에 해당됨
                
                stock_list_ret.append(info[0][0])
                stock_list_who.append(info[0][1])
                stock_list_dis.append(info[0][2])
                stock_list_man.append(info[0][3])
            
                demand_list_ret.append(info[1][0])
                demand_list_who.append(info[1][1])
                demand_list_dis.append(info[1][2])
                demand_list_man.append(info[1][3])
                
                action_real_list_ret.append(info[2][0][-1])
                action_real_list_who.append(info[2][1][-1])
                action_real_list_dis.append(info[2][2][-1])
                action_real_list_man.append(info[2][3][-1])

                rewards_list_ret.append(reward[0])
                rewards_list_who.append(reward[1])
                rewards_list_dis.append(reward[2])
                rewards_list_man.append(reward[3])
            
            
                self.env.render(episode, './실험결과/{}/test_log.txt'.format(PATH))

                # 보상을 부여하고 episode의 종료 판정 및 state_next를 설정한다
                if done:
                    state_next = None  # 다음 상태가 없으므로 None으로
                    break

                else:
                    state_next = observation_next  # 관측 결과를 그대로 상태로 사용
                    state_next = torch.tensor(state_next).type(torch.FloatTensor).cuda()
                    state_ret_next = state_next[:,:5].reshape(-1).unsqueeze(0)
                    state_who_next = state_next[:,5:10].reshape(-1).unsqueeze(0)
                    state_dis_next = state_next[:,10:15].reshape(-1).unsqueeze(0)
                    state_man_next = state_next[:,15:].reshape(-1).unsqueeze(0)

                # 관측 결과를 업데이트
                state = state_next
                
            if episode == 0 :
                if '테스트' not in os.listdir('./실험결과/{}'.format(PATH)) :
                    os.mkdir('./실험결과/{}/테스트'.format(PATH))

                plot_test(PATH+'/테스트', -5.71, 1, "Retailer", stock_list_ret, demand_list_ret, action_real_list_ret, rewards_list_ret, episode)
                plot_test(PATH+'/테스트', -8.795, 2, "Wholesaler", stock_list_who, demand_list_who, action_real_list_who, rewards_list_who, episode)
                plot_test(PATH+'/테스트', -12.76, 3, "Distributor", stock_list_dis, demand_list_dis, action_real_list_dis, rewards_list_dis, episode)
                plot_test(PATH+'/테스트', -11.71, 4, "Manufacturer", stock_list_man, demand_list_man, action_real_list_man, rewards_list_man, episode)
            else :
                plot_test(PATH, -5.71, 1, "Retailer", stock_list_ret, demand_list_ret, action_real_list_ret, rewards_list_ret, episode)
                plot_test(PATH, -8.795, 2, "Wholesaler", stock_list_who, demand_list_who, action_real_list_who, rewards_list_who, episode)
                plot_test(PATH, -12.76, 3, "Distributor", stock_list_dis, demand_list_dis, action_real_list_dis, rewards_list_dis, episode)
                plot_test(PATH, -11.71, 4, "Manufacturer", stock_list_man, demand_list_man, action_real_list_man, rewards_list_man, episode)
```
